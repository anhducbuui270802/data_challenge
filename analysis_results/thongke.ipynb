```python
def converdf(data):
    sample = []
    for id, item in data.items():
        data_sample = {}
        data_sample['id'] = id
        data_sample['evidence'] = item['evidence']
        data_sample['verdict'] = item['verdict']
        data_sample['evidence_predict'] = item['evidence_predict']
        data_sample['idx_evidence'] = item["idx_evidence"]
        data_sample['percentages_rationale'] = item['percentages_rationale']
        sample.append(data_sample)
        try:
            data_sample['verdict_predict'] = item['verdict_predict']
            data_sample['percentages_labels'] = item['percentages_labels']

        except:
            pass

    return pd.DataFrame(sample)
```

```python
import re

def tim_tu_co_so(cau):
    pattern = r'\b\w*\d\w*\b'  # Biểu thức chính quy để tìm các từ chứa số
    ket_qua = re.findall(pattern, cau)
    return ket_qua

# Ví dụ sử dụng:
cau = "It's 12 am and I have 3fsd apples."
ket_qua = tim_tu_co_so(cau)
print(ket_qua)
```

```python
def delete_item(l1,l2):
    l1 = l1.split(" ")
    l2 = l2.split(" ")
    duplicate = []
    for i in l2:
        if i in l1:
            duplicate.append(i)

    l1 = " ".join([x for x in l1 if x not in duplicate])
    l2 = " ".join([x for x in l2 if x not in duplicate])
    print("str1: ", l1)
    print("str2: ", l2)

    return l1,l2
```
