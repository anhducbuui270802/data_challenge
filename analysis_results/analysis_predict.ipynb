````python
def assign_verdict(df):
    df.loc[df.pred==0, "pred"] = df.loc[df.pred==0, "verify_pred"]
    # df.loc[(df.pred==2) & (df.prob<=0.55) & (df.verify_prob <=0.6), "pred"] = 2
    # df.loc[(df.pred==2) & (df.prob<=0.55) & (df.verify_prob <0.7), "pred"] = df.loc[(df.pred==2) & (df.prob<=0.55) & (df.verify_prob <0.7), "verify_pred"]
    # df.loc[(df.pred==2) & (df.prob<=0.55) & (df.verify_prob >0.6), "pred"] = df.loc[(df.pred==2) & (df.prob<=0.55) & (df.verify_prob >0.6), "verify_pred"]
    # df.loc[(df.pred==2) & (df.prob>0.55), "pred"] = 2
    df.loc[(df.pred==2) & (df.prob>0.55), "pred"] = 2
    return df
````

````python
verdict_ids = []
verdicts = []
verdict_evidence = []
for idx in test_root_df.index:
    verdict_ids.append(idx)

    tmp = copy_df[copy_df.id==idx]
    if 1 in tmp.pred.values:
        sentences = tmp[tmp.pred==1].sentence_root.values
        verdict_evidence.append(sentences[0])
        verdicts.append("REFUTED")

    elif 0 in tmp.pred.values:
        sentences = tmp[tmp.pred==0].sentence_root.values
        verdict_evidence.append(sentences[0])
        verdicts.append("SUPPORTED")
        
    else:
        verdict_evidence.append("")
        verdicts.append("NEI")
````

````python
data = {
    "verdict" : verdicts,
    "evidence" : verdict_evidence
}
````

````python
# pd.set_option("display.max_rows", None)
results = pd.DataFrame(data=data)
````

````python
from sklearn.metrics import matthews_corrcoef,accuracy_score, f1_score
````

````python
def acc_and_f1(preds, labels):
    acc = accuracy_score(y_true=labels, y_pred=preds)
    f1 = f1_score(y_true=labels, y_pred=preds, average="macro")
    return {
        "acc": acc,
        "f1": f1,
        "acc_and_f1": (acc + f1) / 2,
    }
````
